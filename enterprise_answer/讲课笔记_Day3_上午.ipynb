### Cell b0 - 什么是 RAG？（核心概念讲解）

| 项目 | 内容 |
|:---|:---|
| **时间** | 0:12-0:20，8分钟 |
| **对应Cell** | b0（markdown） |
| **操作** | 纯讲解，不需要运行 |

**核心概念：RAG 解决 LLM 的三个根本性限制**

**讲解要点（这是全天最重要的概念之一，要讲透）：**

> "RAG，全称 Retrieval-Augmented Generation，检索增强生成。名字很学术，但原理非常直觉。"
>
> "**LLM 像一个很聪明但不看最新新闻的人。RAG 就是给他一本最新的参考书。**"
>
> "为什么需要 RAG？因为 LLM 有三个'硬伤'，不是优化能解决的，是架构性的限制："
>
> "**第一，知识截止**。GPT-4 的训练数据截止到某个时间点，之后的事情它完全不知道。你问它'2025年发生了什么'，它只能猜。RAG 怎么解决？给它一个实时更新的知识库，每次回答前先查一下最新信息。"
>
> "**第二，幻觉**。LLM 最危险的特点就是——它不知道自己不知道什么。它会非常自信地编造一个看起来很对的答案。比如你问它一个很冷门的问题，它不会说'我不知道'，它会编一个答案。RAG 怎么解决？让它基于真实文档回答，而且可以标注来源。"
>
> "**第三，没有领域知识**。LLM 不了解你们公司内部的数据、你们行业特有的知识。你不可能把公司内部文档都拿去训练一个模型。RAG 怎么解决？建一个企业私有知识库，LLM 每次回答前先检索。"

**要指出的重点：**
- Cell 中的流程图：用户提问 -> 向量化 -> 搜索 -> 拼接 -> LLM 生成
- 表格中三个限制和对应的 RAG 解决方案
- 强调这是**架构性**的解决方案，不是 prompt 技巧

**常见学生问题：**
- Q: "RAG 和直接把文档放在 prompt 里有什么区别？" A: "好问题！直接放 prompt 有 context window 限制（比如 128K token），而且不知道该放哪些内容。RAG 的检索步骤就是帮你自动挑选最相关的内容。"
- Q: "RAG 能完全消除幻觉吗？" A: "不能完全消除，但能大大减少。而且可以提供来源引用，让用户能验证。练习2我们会专门测试 RAG 的失败模式。"
- Q: "向量化是什么意思？" A: "就是把文本变成一个数字向量，Day1 讲过的 Embedding。语义相似的文本，向量也相似。等下我们就会看到。"

**过渡：** "概念讲完了，接下来我们真的动手建一个向量知识库。"

### 练习 1：分块策略设计（Cell 17baa4be）

| 项目 | 内容 |
|:---|:---|
| **时间** | 0:34-0:45，约11分钟（讲解3分钟 + 动手5分钟 + 讲解结果3分钟） |

| **操作** | 学生动手，然后运行参考答案 |

**核心概念：Chunking 策略对检索质量的影响**

**练习前讲解（3分钟）：**

> "在实际项目中，文档不可能像刚才那样一两句话。一篇公司制度文档可能有几千字。直接把整篇文档做 embedding？太长了，语义太分散。所以需要**分块（chunking）**。"
>
> "分块策略是 RAG 系统中**最影响效果的环节之一**。分得太细，丢失上下文；分得太粗，语义太杂。"
>
> "这个练习给你一段公司差旅报销制度的文本（约300字），要你实现两种分块："
> - **按句分块**：用句号分割，每个 chunk 是一句话
> - **按段分块**：用空行分割，每个 chunk 是一个段落
>
> "然后对比搜索'报销流程'时，哪种分块效果更好。"
>
> "给大家 5 分钟时间。不会的可以打开提示（hints）。"

**学生动手期间巡场要点：**
- 确认学生理解了 `re.split(r'[。！？]', text)` 语法
- 确认学生理解了 `text.split('\n\n')` 语法
- 如果有人很快做完，提示他思考：为什么段落分块可能更好？

**运行后要看的输出（讲解结果 3分钟）：**

> "看结果："
> - 方案A（按句分块）：**10 个 chunks**
> - 方案B（按段分块）：**4 个 chunks**
>
> "搜索'报销流程是什么？'的结果："
> - 方案A top1 相关度 **0.772**
> - 方案B top1 相关度 **0.761**
>
> "相关度差不多对吧？但关键区别在于**检索到的内容**："
> - 方案A 只检索到一句话：'出差结束后5个工作日内需提交报销申请'
> - 方案B 检索到整个段落：包含申请、发票、审批金额、打款时间等完整信息
>
> "**结论：段落分块更好**，因为它保留了完整的上下文。单句太碎片化了，虽然相关度分数可能略高，但信息量不够，LLM 基于一句话没法生成有用的回答。"
>
> "这就是实际 RAG 项目中的核心权衡：**precision vs. context completeness**。"

**延伸讨论：**

> "实际项目中的分块策略会更复杂："
> - **滑动窗口分块**：固定大小但有重叠（overlap），避免在关键信息中间切断
> - **语义分块**：用 NLP 判断语义边界
> - **递归分块**：LangChain 的 RecursiveCharacterTextSplitter

**常见问题：**
- Q: "最佳的 chunk 大小是多少？" A: "没有标准答案。一般 200-500 字比较常见。需要根据你的文档类型和查询类型实验。"
- Q: "metadata 有什么用？" A: "可以做混合检索——先用 metadata 过滤（比如只搜'报销'类文档），再在子集里做向量搜索。"

**过渡：** "好，分块和检索都搞定了。接下来我们把检索和 LLM 生成连起来，构建完整的 RAG 系统。"

| 项目 | 内容 |
|:---|:---|

| **对应Cell** | ReActAgent 类定义 |
| **操作** | 运行 Cell，逐段讲解代码逻辑 |

**核心概念：ReAct = Reasoning + Acting，Thought-Action-Observation 循环**

**讲解要点（这是下半场最重要的代码，要讲透）：**

> "ReAct 是 2022 年一篇论文提出的 Agent 范式。名字就是 Reasoning + Acting。核心思路："
>
> "LLM 每一步要做三件事："
> 1. **Thought（思考）**：我需要做什么？
> 2. **Action（行动）**：调用哪个工具？
> 3. **Observation（观察）**：工具返回了什么？
>
> "然后基于观察结果，继续思考，直到有足够信息给出 **Final Answer**。"
>
> "看代码结构："
> - `run()` 方法是主循环，最多跑 `max_steps`（默认5步）
> - 每一步：构造 prompt（包含工具描述和历史记录），让 LLM 生成
> - 解析 LLM 输出：如果有 'Final Answer:' 就返回答案
> - 如果有 'Action:'，提取工具名和参数，调用工具
> - 把工具结果作为 'Observation:' 加入历史，继续下一步

**要特别指出的代码细节：**

> "prompt 的格式非常重要。我们告诉 LLM 必须严格按这个格式输出："
> ```
> Thought: 我需要...
> Action: tool_name
> Action Input: {"param1": "value1"}
> ```
> "或者："
> ```
> Thought: 我已经有足够信息
> Final Answer: ...
> ```
>
> "这个格式化的要求就是让 LLM 的输出可以被程序解析。如果 LLM 不按格式输出，Agent 就没法工作。这也是为什么 Agent 需要比较强的 LLM——它需要严格遵循指令格式。"

**常见问题：**
- Q: "max_steps 为什么是 5？" A: "防止无限循环。实际项目中可以根据任务复杂度调整。如果 5 步之内没搞定，可能是查询太复杂或工具不够用。"
- Q: "这跟 OpenAI Function Calling 有什么区别？" A: "Function Calling 是 API 层面原生支持的工具调用。ReAct 是 prompt 层面的方案，更通用，任何 LLM 都能用。"

**运行后要看的输出：**
- `[OK] ReAct Agent 已创建`

>
> "**代码生成**——**微调最佳**。代码的模式比较固定，微调能让模型学会你们公司的代码风格、框架用法。RAG 在这个场景效果不好，因为代码片段的语义检索不够精确。"
>
> "**数据分析**——**两者结合最佳**。微调让模型理解你们的数据表结构和业务逻辑，RAG 提供实时数据。"

**决策框架（重点！用表格总结）：**

> "简单记忆："
> - **知识更新频率高** -> RAG（不需要重新训练）
> - **需要特定风格/格式** -> 微调（让模型学会你的'味道'）
> - **需要可追溯性** -> RAG（可以引用来源）
> - **数据量大且结构化** -> 微调（内化知识效率更高）
> - **预算有限** -> RAG（不需要 GPU 训练）
> - **两者都需要** -> 微调 + RAG（先微调基础能力，再 RAG 补充实时知识）

**常见问题：**
- Q: "我们公司应该用哪个？" A: "大多数企业场景建议先从 RAG 开始。成本低、见效快、迭代方便。如果 RAG 效果不够好，再考虑加微调。"
- Q: "RAG 的成本是多少？" A: "主要成本是 embedding API 调用（建库时一次性）+ LLM API 调用（每次查询）。比微调便宜很多。"
