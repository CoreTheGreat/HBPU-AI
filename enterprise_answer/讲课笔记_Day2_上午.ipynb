## Part 1：LLM 生命周期全景（Cell 3）

**时间分配：12 分钟**

**要运行的 Cell：Cell 3（Markdown，不需要运行，滚动讲解）**

### 关键概念
大模型从出生到上岗的 4 个阶段：Pretrain -> SFT -> Alignment -> Deploy

### 讲话要点

> 大家看这张表。一个大语言模型从"出生"到"上岗"，要经过 4 个阶段。我一个一个讲。
>
> **第一阶段：预训练（Pretraining）**
> - 干什么？给模型喂 TB 级别的文本——互联网、书籍、代码、百科——让它学会语言的统计规律。
> - 产出什么？一个 Base Model。它"什么都知道一点"，但它只会一件事：**给定前文，预测下一个字**。
> - 企业自己做吗？**几乎不做**。预训练要几百张 GPU 跑几个月，花几百万人民币。我们用开源的就行。
>
> **第二阶段：SFT（指令微调）**
> - 干什么？用万级到十万级的 instruction-response 数据对，教模型"你问什么我就答什么"。
> - 产出什么？一个 Chat Model。它能对话了。
> - 这是企业**最常见的切入点**！今天下午的 LoRA 也属于 SFT 的一种高效做法。
>
> **第三阶段：对齐（Alignment）**
> - 用 RLHF/DPO 等方法，让模型输出更安全、更符合人类偏好。
> - 明天上午会讲 DPO。
>
> **第四阶段：部署（Deployment）**
> - 量化、缓存、API 化。明天下午会讲。
>
> **关键认知**（指着屏幕上的"关键认知"部分）：
> - 预训练给模型"知识"，SFT 给模型"行为"。
> - 打个比方：预训练像让一个人读了一辈子书，SFT 像培训他当客服。读了很多书不代表他会当客服，但当客服必须先有知识储备。
> - 企业落地路径：**拿开源 Base Model -> SFT 适配业务场景**。这就是我们今天要走的路。

### 要指出的重点
- 表格中"企业通常从哪进入"那一列——SFT 行写着"最常见切入点"，要用手指着强调
- 预训练的"很少自己做"——帮学员建立正确预期

### 常见问题

- **Q：SFT 和 Fine-tuning 有什么区别？**
  - A：SFT 是 Fine-tuning 的一种。Fine-tuning 是泛称，SFT 特指用 instruction-response 格式数据做微调。还有其他 fine-tuning 方式比如 RLHF。

- **Q：企业真的不做预训练吗？**
  - A：99% 不做。除非你是百度、阿里级别的公司，有专门的预训练团队。中小企业直接用 Llama、Qwen 等开源模型做 SFT 就够了。

- **Q：对齐（Alignment）和 SFT 的区别是什么？**
  - A：SFT 教模型"怎么做"，对齐教模型"什么该做什么不该做"。SFT 用的是明确答案，对齐用的是"好/坏"的偏好对比。明天会详细讲。

## Part 2：预训练 Demo — 模型定义（Cell 12-14）

**时间分配：5 分钟**

**要运行的 Cell：Cell 12（Markdown）、Cell 13（模型代码）、Cell 14（训练前生成）**

### 关键概念
4 层 Transformer 模型结构 + 理论初始 loss = ln(vocab_size)

### 讲话要点

> - 4 层 Transformer
> - 128 维 embedding
> - 4 个 attention head
> - 总参数量大约 **848,384 个**，约 **3.24MB**

> 做个对比：GPT-3 有 1750 亿参数，我们这个是它的**五十万分之一**。但麻雀虽小，五脏俱全——核心组件完全一样。
>
> 运行 **Cell 13** 定义模型。代码比较长，不用逐行看，关键是理解它有这几个组件：
> - Token Embedding：把每个字映射成 128 维向量
> - Position Embedding：编码位置信息

>
> 然后运行 **Cell 14**，这个很重要！它做了两件事：
>
> 第一，计算**理论初始 loss**：
> - 随机初始化的模型，对 366 个字符均匀猜，理论 loss = ln(366) = **5.9026**
> - 这个数字是我们的**基线**！训练后 loss 应该远低于它。
>
> 第二，用未训练的模型生成文本——**输出是纯粹的乱码**。
> - 大家看输出，是不是完全不通顺？像猴子在键盘上乱打。

### 要指出的输出

- 理论初始 loss = ln(366) = 5.9026——在白板上写下来，后面对比用
- 训练前的生成文本——纯乱码，让学员印象深刻

### 常见问题

- **Q：为什么理论 loss 是 ln(vocab_size)？**
  - A：随机模型对 366 个字符均匀分配概率，每个字概率 = 1/366。Cross-entropy = -log(1/366) = ln(366)。如果实际初始 loss 接近这个值，说明模型初始化是合理的。

- **Q：为什么模型这么小？真实模型多大？**
  - A：教学演示用。真实模型从几亿到几千亿参数不等。后面 SFT 部分我们用的 GPT-2 Chinese 有 1 亿多参数，已经大多了。

## Part 3：ChatML 格式 + Loss Masking（Cell 34-36）

**时间分配：15 分钟**

**要运行的 Cell：Cell 34（Markdown 概念讲解）、Cell 35（ChatML 格式演示）、Cell 36（Loss Masking 可视化）**

### 关键概念
SFT 的两个核心概念：ChatML 对话格式 + Loss Masking（只在 assistant 回复部分计算 loss）

### 这是整个上午最重要的 15 分钟！

### 讲话要点

> 现在我要讲 SFT 最核心的两个概念。如果今天上午只记住一个东西，就记住这个。
>
> **第一个概念：ChatML 格式**
>
> 看 **Cell 34** 的 Markdown。ChatML 是一种对话标记语言，让模型知道"谁在说话"：
> ```
> <|im_start|>user
> 今天天气怎么样？<|im_end|>
> <|im_start|>assistant
> 今天天气晴朗，适合出门。<|im_end|>
> ```
> 这些特殊标记（`<|im_start|>`, `<|im_end|>`）是分隔符，让模型区分 user 和 assistant 的部分。
>
> 运行 **Cell 35** 看实际效果。
>
> **第二个概念：Loss Masking——这才是 SFT 的核心 trick！**
>
> 大家想一个问题：SFT 训练时，模型应该学习什么？
> - 学习 user 的提问？**不！** user 的提问是输入，不是模型要生成的。
> - 学习 assistant 的回复？**对！** 我们只想让模型学会"怎么回答"。
>
> 所以 Loss Masking 的做法是：
> ```
> input_ids:  [用户问题的tokens...] [助手回复的tokens...] [EOS]
> labels:     [-100 -100 -100 ... ] [助手回复的tokens...] [EOS]
> ```
> - user 部分的 labels 全部设为 **-100**
> - PyTorch 的 CrossEntropyLoss 遇到 -100 就自动跳过，不算 loss、不算梯度
> - 只有 assistant 部分参与训练
>
> 运行 **Cell 36** 看可视化效果。输出是一个表格：
> - 每一行是一个 token
> - `[MASK]` 表示这个位置不算 loss（user 部分）
> - `[TRAIN]` 表示这个位置算 loss（assistant 部分）
>
> 注意输出的统计信息——大部分 token 被 MASK，只有少数 TRAIN。
>
> 看到没？**大部分 token 都不参与训练！** 只有 assistant 的回复部分才算 loss。这就是为什么 SFT 比预训练高效得多——它只需要学"怎么回答"，不需要重新学所有语言知识。

### 要指出的输出
- Cell 35：ChatML 格式的实际文本——强调 `<|im_start|>` 和 `<|im_end|>` 标记
- Cell 36 的表格：每个 token 的 MASK/TRAIN 状态
- Cell 36 的统计：MASK 比例约 68-74%——大部分 token 不参与训练
- **-100 这个魔法数字**——PyTorch 的约定，遇到 -100 就跳过

### 常见问题

- **Q：为什么不让模型也学 user 的提问部分？**
  - A：因为 user 的提问是千变万化的，模型不需要"学会提问"。它只需要学会"给定提问，怎么回答"。如果也学提问，反而会干扰训练效果。

- **Q：-100 是怎么回事？可以用其他值吗？**
  - A：-100 是 PyTorch CrossEntropyLoss 的 `ignore_index` 默认值。你可以改成其他值，但需要同步修改 loss 函数的参数。用 -100 是行业惯例。

- **Q：ChatML 是唯一的对话格式吗？**
  - A：不是。还有 Alpaca 格式、Vicuna 格式、Llama 格式等。但核心思想一样：区分角色 + 只训练 assistant 部分。ChatML 是目前最流行的格式之一。

- **Q：multi-turn（多轮对话）怎么处理？**
  - A：原理一样。多轮对话有多个 user-assistant 交替，所有 user 部分都 mask，所有 assistant 部分都参与训练。

### 常见问题

- **Q：为什么 SFT 的学习率要比预训练小？**
  - A：因为预训练已经给模型建立了知识基础，SFT 只是"微调"。学习率太大会破坏预训练的知识（叫做"灾难性遗忘"）。一般 SFT 的学习率是预训练的 1/10 到 1/100。

- **Q：100 步够吗？**
  - A：对于这个任务够了！因为任务相对简单（6 分类），数据质量高，而且模型已经有了预训练知识。复杂任务可能需要更多步。

- **Q：loss 在 step 80 降到 0.05 然后又回升到 0.07，是过拟合吗？**
  - A：小幅波动是正常的。如果 loss 持续上升才需要担心。我们的 val loss 仍然很低（0.0356），说明没有过拟合。

> - 太小（1e-5）：学得慢，可能还没学会就停了
> - 太大（2e-4）：学得猛，但可能"忘掉"预训练知识

>

>
> 企业启示：超参调优是 SFT 落地的必经之路。不要"一把梭"用默认参数。至少做一轮学习率搜索。

### 常见问题

- **Q：除了学习率，还有什么重要的超参数？**
  - A：batch_size、epochs/steps、warmup ratio、weight decay。其中 batch_size 和 epochs 也很重要。一般做法是先搜 lr，再搜 epochs，最后微调其他参数。

- **Q：50 步够做对比吗？**
  - A：做初步筛选够了。找到大致的 best lr 后，再用这个 lr 跑更多步做精细训练。实际企业中也是分两步：粗搜 + 精搜。

- **Q：2e-4 效果差是因为"灾难性遗忘"吗？**
  - A：是的。学习率太大，模型更新太猛，会破坏预训练时学到的语言知识。LoRA 技术（下午讲）可以缓解这个问题——它只更新很少的参数，不容易遗忘。
