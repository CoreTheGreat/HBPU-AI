## 🛰️ 阿里云百炼 (DashScope) API 调用说明

本节介绍从注册到获取 API Key 的完整流程。请根据步骤完成操作后补充截图。

1. 注册阿里云账号并登录：
   - 注册的时候不要急，可能有学生优惠的羊毛可以薅
   - 羊毛是动态的，老师给不了指导，只能靠你们雪亮的眼睛了
   - 访问：https://www.aliyun.com
   - 如果已有账号可直接登录

2. 进入 DashScope 控制台：
   - 访问：https://bailian.console.aliyun.com/
   - 选择"应用 -> API-Key"
   
   ![image.png](images/env_val_01.png)

3. 创建 API Key：
   - 点击 "创建 API Key"
   - 复制 `API-Key`（一般为OpenAI兼容格式：`sk-xxxx`）

   ![image.png](images/env_val_02.png)

4. Token额度查询：
   - 点击 "模型 -> 模型用量
   - 有20根左右的羊毛可薅，1,000,000 Tokens/个

   ![image.png](images/env_val_03.png)

```python
# 阿里云百炼 API 验证（OpenAI 兼容接口示例）

import os
from openai import OpenAI

# ============================================================
# 👇 把你的 API Key 填在下面的引号里
# ============================================================
DASH_SCOPE_API_KEY = "" # 例如: "sk-abc123..."
# 
try:
    client = OpenAI(
        # 若没有配置环境变量，请用阿里云百炼API Key将下行替换为: api_key="sk-xxx",
        api_key=DASH_SCOPE_API_KEY,
        base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
    )

    response = client.chat.completions.create(
        model="qwen3.5-plus",  # 模型列表: https://help.aliyun.com/model-studio/getting-started/models
        messages=[
            {'role': 'system', 'content': 'You are a helpful assistant.'},
            {'role': 'user', 'content': '你是谁？'}
        ]
    )
    print(response.choices[0].message.content)

except Exception as e:
    print(f"错误信息：{e}")
    print("请参考文档：https://help.aliyun.com/model-studio/developer-reference/error-code")
```

## 🔍 Embedding 后端测试

测试阿里云百炼的 Embedding 模型是否可用。

![image.png](images/env_val_04.png)

```python
# Embedding 后端测试（使用阿里云百炼 text-embedding-v3）

import os
from openai import OpenAI

try:
    client = OpenAI(
        api_key=DASH_SCOPE_API_KEY,
        base_url="https://dashscope.aliyuncs.com/compatible-mode/v1",
    )

    response = client.embeddings.create(
        input="Hello world",
        model="text-embedding-v4"  # 模型列表: https://help.aliyun.com/model-studio/getting-started/models
    )
    embedding = response.data[0].embedding
    print(f"✅ Embedding 后端正常，维度: {len(embedding)}")
    print(f"示例向量前5个值: {embedding[:5]}")

except Exception as e:
    print(f"❌ Embedding 后端异常: {e}")
    print("请参考文档：https://help.aliyun.com/model-studio/developer-reference/error-code")
```

3. 模型选择：
   - 进入模型广场
   - 查看模型的Token价格，记录 model (chat.completions.create方法的参数)

   ![image.png](images/env_val_07.png)

```python
# 硅基流动 API 验证（OpenAI 兼容接口示例）

from openai import OpenAI

# ============================================================
# 👇 把你的 SiliconFlow API Key 填在下面的引号里
# ============================================================
SILICONFLOW_API_KEY = ""  # 例如: "sk-abc123..."

try:
    client = OpenAI(
        api_key=SILICONFLOW_API_KEY,
        base_url="https://api.siliconflow.cn/v1",
    )

    # 测试 LLM
    response = client.chat.completions.create(
        model="deepseek-ai/DeepSeek-V3.2",  # 模型列表: https://siliconflow.cn/models
        messages=[
            {'role': 'system', 'content': 'You are a helpful assistant.'},
            {'role': 'user', 'content': '你是谁？'}
        ]
    )
    print("✅ SiliconFlow LLM 后端正常")
    print(response.choices[0].message.content)

except Exception as e:
    print(f"❌ SiliconFlow API 异常: {e}")
    print("请参考文档：https://siliconflow.cn/docs")
```

```python
try:
    # 测试 Embedding
    response_emb = client.embeddings.create(
        input="Hello world",
        model="BAAI/bge-m3"  # 模型列表: https://siliconflow.cn/models
    )
    embedding = response_emb.data[0].embedding
    print(f"✅ SiliconFlow Embedding 后端正常，维度: {len(embedding)}")
    print(f"示例向量前5个值: {embedding[:5]}")

except Exception as e:
    print(f"❌ SiliconFlow API 异常: {e}")
    print("请参考文档：https://siliconflow.cn/docs")
```
